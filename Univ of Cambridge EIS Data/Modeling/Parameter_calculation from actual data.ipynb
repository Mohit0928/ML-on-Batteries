```python
R0=[]
R1_R2=[]
im_max=[]
slope=[]
for i in range(262):
    freq=df[:,2][df[:,1]==1]
    real=df[:,3][df[:,1]==i+1]
    img=(df[:,4][df[:,1]==i+1])*1j
    abs_Z=df[:,5][df[:,1]==i+1]


    for j in range(60):

        Z[j]=real[j]-img[j]

    Z=Z[:60]
    from impedance.models.circuits import CustomCircuit as cc
    circuit = 'R0-p(R1,C1)-p(R2-Wo1,C2)'
    initial_guess = [.45, .01, 100, .01, .05, 100, 1]
    circuit=cc(circuit=circuit,initial_guess=initial_guess)
    circuit=circuit.fit(freq,Z)
    Z_fit=circuit.predict(freq)
    
#     import matplotlib.pyplot as plt
#     from impedance.visualization import plot_nyquist

#     fig, ax = plt.subplots()
#     plot_nyquist(ax, Z, fmt='o')
   

# #     plt.legend(['Data', 'Fit'])
#     plt.show()
    k=0
    for j in img:
        if j>0:
            R0.append(real[k])
            k+=1
            break
    imag_min=img[35:50]
    imag_max=img[50:]
    ind_min=int(np.argmin(imag_min))
    ind_max=int(np.argmax(imag_max))
    
    R1_R2.append(real[ind_min+35]-R0[i])
   
    slope.append((abs(img[ind_max+50])-abs(img[ind_min+35]))/real[ind_max+50]-real[ind_min+35])
    
    im_max.append(abs(max(img[20:35])))
    
    
```
