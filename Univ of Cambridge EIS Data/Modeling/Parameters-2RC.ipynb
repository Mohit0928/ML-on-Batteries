```python
df=np.genfromtxt("EIS_state_V_25C01.txt",delimiter="\t")
fr,Z = preprocessing.readCSV('./exampleData.csv')
R0_25c01=[]
R1_25c01= []
C1_25c01 = []
R2_25c01 = []
C2_25c01 = []
Wo1_0_25c01 = []
Wo1_1_25c01 = []
```

```python

n=261
for k in range(1,n+1):
    if(k==1):
        freq=df[:,2][df[:,1]==1]
        real=df[:,3][df[:,1]==k]
        img=(df[:,4][df[:,1]==k])*1j
    
    else:        
        freq=df[:,2][df[:,1]==1]
        real=df[:,3][df[:,1]==k-1]
        img=(df[:,4][df[:,1]==k-1])*1j


    for i in range(np.size(freq)):

        Z[i]=real[i]-img[i]
    Z=Z[:60]
    from impedance.models.circuits import Randles,CustomCircuit as cc
    circuit = 'R0-p(R1,C1)-p(R2-Wo1,C2)'
    initial_guess = [.01, .01, 100, .01, .05, 100, 1]
#     circuit = 'R0-p(R1,C1)-p(R2,C2)-p(R3-Wo1,C3)'
#     initial_guess = [.01, .01, 100, .01, 100,.01, .05, 100, 1]
    circuit=cc(circuit=circuit,initial_guess=initial_guess)
    circuit=circuit.fit(freq,Z)
    
    R0_25c01.append(circuit.parameters_[0])
    R1_25c01.append(circuit.parameters_[1])
    C1_25c01.append(circuit.parameters_[2])
    R2_25c01.append(circuit.parameters_[3])
    C2_25c01.append(circuit.parameters_[6])
#     R3_25c01.append(circuit.parameters_[5])
    Wo1_0_25c01.append(circuit.parameters_[4])
    Wo1_1_25c01.append(circuit.parameters_[5])
#     C3_25c01.append(circuit.parameters_[8])
    
   
```
