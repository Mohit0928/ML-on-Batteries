# Approach 1

In this approach I used Random Forest Regression with mean, variance, minimum of change in
discharge voltage curves between the 100th and 1st cycle: ∆Q100−1 (V ) = Q100 (V ) - Q1 (V ) as features. On short life cells, I get an MAE of 6.5% and 8.5% on long-life cells, making an aggregate of
(22x6.5%+18x8.5%)/40=7.4%. When I use all training set without categorizing into short and long life
cells, I get an error of 13.5% on secondary set. This shows that choosing a correct dataset for a model
is an essential part. Comparing my results with Severson datasets, they have got an MAE of 9.5%, so
there is a decrease of 2.1% error. While looking at individual categories, I am getting an MAE of only
6.5% for short life cells. Moreover my model can make continuous prediction till its EOL, when I use
mean, variance, minimum of change in discharge voltage curves between the i th and 1 st cycle: ∆Q i−1 (V )
= Q i (V ) - Q 1 (V ), for i>=100 as feature. MAE decreases as the i in ∆Qi−1 (V ) increases. When I use
a different ML algorithm (Lasso) for short life cells training and testing with the same features, the error
decreases to 5%. This further decreases the aggregate MAE to (22x5.2%+18x8.5%)/40= 6.7%

## Short Life Cells

```python
from scipy.stats import kurtosis
from scipy.stats import skew
var_train=[]
var_test=[]
mean_train=[]
mean_test=[]
min_train=[]
min_test=[]
skew_train=[]
skew_test=[]
kurtosis_train=[]
kurtosis_test=[]
j=0
a=0
y=[]
for key in curves:
    a=int((curves[key].size)/100-3)
    j+=1
    
    if (a>1000):                  #Remove cells with EOL greater than 1000
        continue
    
    
    if  j>84:                      # Batch 3 cell for testing
        y.append(a)
        a=(curves[key]["Q_"+ str(100)]-curves[key]["Q_"+ str(1)])         # Q10(V ) - Q1 (V )

        var_test.append(np.log(np.abs(np.var(a))))
        mean_test.append(np.log(np.abs(np.mean(a))))
        min_test.append(np.log(np.abs(np.min(a))))
        skew_test.append(np.log(np.abs(skew(a))))
        kurtosis_test.append(np.log(np.abs(kurtosis(a))))
        
        
            
    if j<42:                       #Batch 1 cell for training
        y.append(a)
        a=(curves[key]["Q_"+ str(100)]-curves[key]["Q_"+ str(1)])    # Q10(V ) - Q1 (V )

        var_train.append(np.log(np.abs(np.var(a))))
        mean_train.append(np.log(np.abs(np.mean(a))))
        min_train.append(np.log(np.abs(np.min(a))))
        skew_train.append(np.log(np.abs(skew(a))))
        kurtosis_train.append(np.log(np.abs(kurtosis(a))))
      
    

x_train_1=pd.DataFrame({"mean":mean_train,"var":var_train,"min":min_train,"skew":skew_train,"kurtosis":kurtosis_train})
x_test_1=pd.DataFrame({"mean":mean_test,"var":var_test,"min":min_test,"skew":skew_test,"kurtosis":kurtosis_test})



y1=y[:int(np.shape(x_train_1)[0])]
y2=y[int(np.shape(x_train_1)[0]):]
y_train=[]
y_test=[]

for i in y1:
    for j in range(int(i),int(i)-1,-1):
        y_train.append(np.log(j))
for i in y2:
    for j in range(int(i),int(i)-1,-1):
        y_test.append(np.log(j))
        
x_train_1["log_cycle_life"]=y_train
x_test_1["log_cycle_life"]=y_test
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

num_pipeline = Pipeline([
    ("std_scaler", StandardScaler())
])

var_attribs = ["var", "min",'mean']
var_pipeline = ColumnTransformer([
    ("num", num_pipeline, var_attribs)
])

import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

def rmse(labels, predictions):
    mse = mean_squared_error(labels, predictions)
    return np.sqrt(mse)

def print_error_and_plot(pipeline, model, train, test):
    fig, ax = plt.subplots()
    ax.plot([0,3000],[0,3000], "k-")
    for name, test_set, marker, color in [
        ("Train", train, "o", "b"), 
        ("Test", test, "^", "y")
    ]:
        X_test = test_set
        y_test = test_set["log_cycle_life"].copy()

        X_test_prepared = pipeline.transform(X_test)
        final_predictions = model.predict(X_test_prepared)
        print("mean_abs_err=",mean_absolute_error(y_test, final_predictions)*100)
        print("{}: {:.0f}".format(name, rmse(np.exp(y_test), np.exp(final_predictions))))
        ax.scatter(np.exp(y_test), np.exp(final_predictions), label=name, marker=marker, color=color)
    ax.legend()
    ax.set_xlabel("Observed cycle life")
    ax.set_ylabel("Predicted cycle life")
    ax.set_xlim([100,1100])
    ax.set_ylim([0,1100])
    plt.show()
    
    return ax
```

 Lasso

```python
from sklearn import linear_model
las = linear_model.Lasso(alpha=.005)
train=var_pipeline.fit_transform(x_train_1)
las.fit(train, y_train)
ax = print_error_and_plot(var_pipeline,las,x_train_1,x_test_1)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor as RFR
train=var_pipeline.fit_transform(x_train_1)
clf_RF = RFR()
clf_RF.fit(train, y_train)
ax = print_error_and_plot(var_pipeline,clf_RF,x_train_1,x_test_1)
```

# Approach 2
In this approach, I have lowered i in ∆Qi−1 (V) from 100 to 10 for short life cells and 40 for long life
cells with the trade of MAE. When I plot variation in capacity with the cell cycles, there are 2 main
observations: 1) The capacity is increasing for initial cycles till 10 th cycle and at 10 th cycle its maximum
as can be seen in table 1 figures. 2) The capacity starts decreasing from 10 th cycle, and then it falls till
40 th cycle, there is local minima at cycle number 40 in the table 2 figures. These two observations give
me the incentive to use them as a feature in our model.
I use mean, variance, and minimum of ∆Q10−1 (V ) as features for short-life cells. This model uses
RFR and gives an MAE of 11.5%, as shown in figure 1a. While for Long life cells, we use mean, variance,
and minimum of ∆Q 40−10 (V ) as features, which gives MAE of 13.85% and RMSE of 211 cycles. Looking
at the RMSE of Long life cells, its nearly the same as in Severson’s paper, and the cells with EOL above
1500 are doing better than the Severson model
